## Visualizing the Data

```python
df
```

```python
#lets find graph between higher age and gender
import matplotlib.pyplot as plt
```

```python
max_1 = df.groupby('Sex')['Age'].max()
min_2 = df.groupby('Sex')['Age'].min()
mean_3 = df.groupby('Sex')['Age'].mean()
```

```python
plt.figure(figsize=(20,10))
plt.bar(max_1.index,max_1.values,label = 'Max Age')
plt.bar(mean_3.index,mean_3.values,label = 'Average Age')
plt.bar(min_2.index,min_2.values,label = 'Min Age')
plt.title("Passenger's Age")
plt.legend()
```

```python
alive = df.groupby('SibSp')['Survived'].max()
```

```python
alive
```

```python
plt.pie(alive.index,alive.values)
plt.legend()
```

```python
Randal = df[df['Survived']==0]
```

```python
Randal
```

```python
alive = Randal.groupby('Sex')['Survived'].count()
```

```python
Raffel = df[df['Survived']==1]
```

```python
Raffel
```

```python
dead = Raffel.groupby('Sex')['Survived'].count()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(20,12))
plt.title('Visual Between Alive Passenger and Dead Passenger')
plt.tight_layout()
plt.subplot(1,2,1)
sns.barplot(x=alive.index,y=alive.values,color='green',label=['Female','Male'])
plt.title('Alive Person')
plt.xlabel('Gender')
plt.ylabel('Count')

plt.tight_layout()


plt.subplot(1,2,2)
sns.barplot(x=dead.index,y=dead.values,color='red',label=['Female','Male'])
plt.title('Dead Person')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.tight_layout()
```
